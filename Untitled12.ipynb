```python
import pandas as pd
import numpy as np
from prophet import Prophet
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('time_series_covid_19_confirmed .csv')
```

```python
# Drop unused cols
df = df.drop(columns=["Lat", "Long"])
```

```python
# Aggregate by country
df = df.groupby("Country/Region").sum().reset_index()
```

```python
# Reshape to long format
df = df.melt(id_vars=["Country/Region"], var_name="Date", value_name="Cases")

```

```python
# Convert to datetime
df["Date"] = pd.to_datetime(df["Date"], format='%m/%d/%y',errors='coerce')
df = df.sort_values(["Country/Region", "Date"])
```

```python
# New daily cases
df['Cases'] = pd.to_numeric(df['Cases'].astype(str).str.replace(' ',''), errors='coerce')
df['Cases']=df.groupby('Country/Region')['Cases'].ffill().fillna(0)
df['NewCases'] = df.groupby('Country/Region')['Cases'].diff().fillna(0)
```

```python
# 7-day moving average
df["Cases_MA7"] = df.groupby("Country/Region")["NewCases"].transform(lambda x: x.rolling(7,min_periods=1).mean())
```

```python
# Growth rate
df["GrowthRate"] = df.groupby("Country/Region")["Cases"].pct_change(fill_method=None).fillna(0)

print(df.head())
```

```python
# استخدام Cases فقط للتجميع
X = df_features[['Cases']].copy()

print("البيانات النهائية للتجميع:")
print(X.describe())
print("\nشكل X:",X.shape)
```

```python
#التجميع النهائي
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. تطبيع البيانات
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_features[['Cases']])

print("بعد التطبيع:")
print("المتوسط:", np.mean(X_scaled, axis=0))
print("الانحراف المعياري:", np.std(X_scaled, axis=0))

# 2. استخدام Elbow Method
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# رسم منحنى Elbow
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('عدد العناقيد')
plt.ylabel('Inertia')
plt.title('Elbow Method لتحديد оптимальное число العناقيد')
plt.grid(True)
plt.show()
```

```python
# 3. التجميع النهائي (افترض أن optimal_k = 3 من الرسم)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df_features['Cluster'] = clusters

print(" تم التجميع بنجاح!")
```

```python
#عرض النتائج
print("=" * 50)
print("نتائج تجميع دول العالم حسب عدد حالات كوفيد-19")
print("=" * 50)

# توزيع العناقيد
cluster_counts = df_features['Cluster'].value_counts().sort_index()
print(f"عدد العناقيد: {optimal_k}")
print("توزيع الدول:")
for cluster_id, count in cluster_counts.items():
    print(f"العنقود {cluster_id}: {count} دولة")

# تحليل كل عنقود
print("\nتحليل كل عنقود:")
for cluster_id in sorted(df_features['Cluster'].unique()):
    cluster_data = df_features[df_features['Cluster'] == cluster_id]

    print(f"\n العنقود {cluster_id} ({len(cluster_data)} دولة):")
    print(f"    متوسط الحالات: {cluster_data['Cases'].mean():.0f}")
    print(f"    أدنى حالة: {cluster_data['Cases'].min():.0f}")
    print(f"    أعلى حالة: {cluster_data['Cases'].max():.0f}")
    print(f"    أمثلة: {cluster_data['Country/Region'].head(5).tolist()}")
```

```python
#النموذج النهائي الجاهز للاستخدام
# نموذج التجميع النهائي
def covid_clustering_model(data, n_clusters=3):
    """
    نموذج تجميع الدول حسب حالات كوفيد-19
    """
    # 1. المعالجة المسبقة
    X = data[['Cases']].copy()
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # 2. التدريب والتنبؤ
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)

    # 3. إضافة النتائج
    data['Cluster'] = clusters
    data['Cluster_Label'] = data['Cluster'].map({
        0: 'منخفض',
        1: 'متوسط',
        2: 'مرتفع'
    })

    return data, kmeans, scaler

# استخدام النموذج
df_final, model, scaler = covid_clustering_model(df_features)
```

```python
#تحليل اداء النموذج
from sklearn.metrics import silhouette_score

# تقييم النموذج
silhouette = silhouette_score(X_scaled, clusters)
print(f" دقة النموذج (Silhouette Score): {silhouette:.3f}")

# تفسير النتيجة:
if silhouette > 0.5:
    print(" التجميع قوي ومتميز")
elif silhouette > 0.3:
    print("  التجميع مقبول ولكن يمكن تحسينه")
else:
    print(" التجميع ضعيف ويحتاج إعادة نظر")
```
